```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = yf.download('SPY', start='2010-01-01', end='2020-12-31').get('Close')
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.SPY)
```

```python
df['return'] = df.SPY.pct_change()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['return'])
```

```python
acf = plot_acf(df['return'].dropna(), lags=30)
```

```python
pacf= plot_pacf(df['return'].dropna())
```
